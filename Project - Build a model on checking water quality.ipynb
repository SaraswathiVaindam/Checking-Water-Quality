```python
### Import packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### Store data from dataset
data = pd.read_csv("water.csv")
data = pd.DataFrame(data)
data
```

```python
data = data.drop(["Potability", "Trihalomethanes", "Sulfate", "Conductivity", "Turbidity", "Organic_carbon"], axis=1).values
df = pd.DataFrame(data)
df.round()
```

```python
df = df.rename({0:"ph", 1:"Hardness", 2:"Solids", 3:"Chloramines"}, axis="columns")
df
```

```python
### Filling NaN values
df = df.bfill()
df.round()
```

```python
x = df.drop(["ph"], axis=1).values
y = df["ph"].values
z = [[x]]
```

```python
### Train the model on the training set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(df[["Hardness", "Solids", "Chloramines"]], df["ph"])
```

```python
### Predicting the y values based on the x values
y_pred = model.predict(df[["Hardness", "Solids", "Chloramines"]])
y_pred
```

```python
### Checking the error
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```
